```python
import pandas as pd

# Load the dataset (use the correct file path)
data = pd.read_excel(r'C:\Users\shara\Downloads\myexcel.xlsx')

# Define age groups
bins = [0, 20, 30, 40, 50, 60, 100]  # Define the age ranges
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '60+']  # Labels for the bins

# Create a new column for age groups
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = data['Age Group'].value_counts()

# Display the predominant age group
predominant_age_group = age_group_counts.idxmax()
print(f"The predominant age group is: {predominant_age_group}")

# Optionally, print the counts of each age group
print(age_group_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (use the correct file path)
data = pd.read_excel(r'C:\Users\shara\Downloads\myexcel.xlsx')

# Calculate the correlation between Age and Salary
correlation = data['Age'].corr(data['Salary'])
print(f"The correlation between Age and Salary is: {correlation}")

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=data)

# Add title and labels
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_excel(r'C:\Users\shara\Downloads\myexcel.xlsx')

# Count of employees in each team
team_counts = data['Team'].value_counts()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=team_counts.index, y=team_counts.values, palette='viridis')
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of employees in each position
position_counts = data['Position'].value_counts()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=position_counts.index, y=position_counts.values, palette='Set2')
plt.title('Segregation of Employees Based on Positions')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define age groups
bins = [0, 20, 30, 40, 50, 60, 100]  # Define the age ranges
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '60+']  # Labels for the bins

# Create a new column for age groups
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Count of employees in each age group
age_group_counts = data['Age Group'].value_counts()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='Blues')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Group by Team and Position and sum the salaries
salary_expenditure = data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Sort by salary expenditure in descending order to get the highest
salary_expenditure = salary_expenditure.sort_values(by='Salary', ascending=False)

# Plotting the top 10 salary expenditures
plt.figure(figsize=(12, 8))
sns.barplot(x='Salary', y='Position', data=salary_expenditure.head(10), hue='Team', palette='coolwarm')
plt.title('Top 10 Team and Position Salary Expenditure')
plt.xlabel('Salary Expenditure')
plt.ylabel('Position')
plt.show()
```

```python
# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=data, color='purple')
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
